# **Aula 02 - Deutsch-Joszua** #


O algoritmo de Deutsch-Josza foi um dos primeiros algoritmos quânticos publicados. Sua importância baseia-se no fato de mostrar que podemos conhecer se a saída de uma função é balanceada (f(0) diferente de f(1)) ou constante (f(0) igual f(1)). Você irá Implemente o algoritmo utilizando o oráculo_1 e depois utilizando o oraculo_2. Cada um dos oraculos utiliza uma funcao f(x) apenas uma vez. Um deles utiliza uma função balanceada, outro, utiliza uma finção constante. Descubra qual é qual utilizando cada um dos oráculos uma vez. 

```python
import numpy as np
from qiskit import *
import matplotlib.pyplot as plt
import matplotlib as mpl
from qiskit.visualization import plot_histogram, plot_bloch_multivector
```

```python
def buildOraculo_1():  
    qc1 = QuantumCircuit(2)
    qc1.x(0)
    qc1.cx(0,1)
    qc1.x(0)
    mygate = qc1.to_gate()
    mygate.label='oraculo 1'
    return mygate
```

```python
def buildOraculo_2():
    qc1 = QuantumCircuit(2)
    qc1.x(1)
    mygate = qc1.to_gate()
    mygate.label='oraculo 2'
    return mygate
```

```python
#inicie seu circuito com 2 qubits quanticos e 1 bit classico
#construa seu algoritmo utilizando um dos oraculos
#para utilizar um dos oraculos, faca: qc.append(buildOraculo_1(),[0,1]) ou qc.append(buildOraculo_2(),[0,1])
#nao esqueca de mostrar a imagem do circuito
qr = QuantumRegister(2, name='q')
cr = ClassicalRegister(1, name='creg')
qc = QuantumCircuit(qr, cr)
qc.x(1)
qc.barrier()
qc.h([0,1])
qc.barrier()

qc.append(buildOraculo_1(), [0,1])
qc.h(0)
qc.measure(0, 0)

qc.draw(output='mpl')
```

```python
#faça diversas medicoes do qubit x e verifique se o oraculo otilizado corresposnde a uma funcao constante ou balanceada.
qasm_sim = Aer.get_backend('qasm_simulator')

shots = 20048

t_qpe = transpile(qc, qasm_sim)

qobj = assemble(t_qpe, shots=shots)

results = qasm_sim.run(qobj).result()

answer = results.get_counts()

plot_histogram(answer)
```

```python
print('A função buildOraculo1 é constate')
```

```python
qr = QuantumRegister(2, name='q')
cr = ClassicalRegister(1, name='creg')
qc = QuantumCircuit(qr, cr)
qc.x(1)
qc.barrier()
qc.h([0,1])
qc.barrier()

qc.append(buildOraculo_2(), [0,1])
qc.h(0)
qc.measure(0, 0)

qc.draw(output='mpl')
```

```python
qasm_sim = Aer.get_backend('qasm_simulator')

shots = 20048

t_qpe = transpile(qc, qasm_sim)

qobj = assemble(t_qpe, shots=shots)

results = qasm_sim.run(qobj).result()

answer = results.get_counts()

plot_histogram(answer)
```

```python
print('A função buildOraculo2 é balançeada')
```
