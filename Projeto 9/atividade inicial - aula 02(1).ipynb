**7)** Vamos agora aprender a construir nossa própria porta quântica, composta por uma sequência de operações com portas conhecidas. Normalmente, construímos nossa porta omo uma função. Observe o código abaixo onde uma porta é criada. 

```python
def constroiMinhaPorta(num_entradas):
    qc1 = QuantumCircuit(num_entradas)
    qc1.x(range(num_entradas))
    mygate = qc1.to_gate().control(2)
    mygate.label='minhaPorta'
    return mygate
```

Peceba que na função acima a porta criada admite ser controlada por 2 qubits! *control(2)*. Voce conseguiria dizer o que essa porta é capaz de fazer e em quais condições atua de fato?  

Vamos agora aprender a utilizar a porta criada. Para isso basta utilizar a função *append*, indicando os bits de controle seguidos dos bits de entrada.

```python
n=4
qr=QuantumRegister(n)
cr=ClassicalRegister(n)
qc=QuantumCircuit(qr,cr)
qc.x(range(2))
qc.barrier()

qc.append(constroiMinhaPorta(int(n/2)),[0,1,2,3])# a ordem dos bits para o appende é sempre promeiro quais sao os de controle e depois os de entrada
qc.barrier()
qc.measure(range(n),range(n))
qc.draw(output='mpl')


```

```python
qasm_sim = Aer.get_backend('qasm_simulator')
shots = 2048
t_qpe = transpile(qc, qasm_sim)
qobj = assemble(t_qpe, shots=shots)
results = qasm_sim.run(qobj).result()
answer = results.get_counts()

plot_histogram(answer)
```
